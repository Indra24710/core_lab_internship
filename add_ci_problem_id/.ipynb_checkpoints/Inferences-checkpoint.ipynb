#### With the learning curve computed based on the accuracy of solving a problem (disregarding the number of submissions made for getting an accepted verdict) for a particular user over time, how does the delta change for a user when the data is aggregated for all the problems?

### Run the below code cell to get the learning curve for a random user. To get the learning curve for a specific user, set 'sel_user' variable to the username of that specific user

```python
#  With the learning curve computed based on the accuracy of solving a problem
# (disregarding the number of submissions made for getting an accepted verdict) 
# for a particular user over time, how does the delta change for a user when the data is aggregated for all the problems?
import matplotlib.pyplot as plt
import random
%matplotlib inline
users=list(set(df_probid['username']))
re=[]
syn=[]
sem=[]
log=[]
sel_user=users[random.randint(0,len(users))]
df_probid_user=df_probid[df_probid['username']==sel_user]
for i in range(len(df_probid_user)):
    re.append(df_probid_user.iloc[i,6:10].sum())
    sem.append(df_probid_user.iloc[i,10:15].sum())
    log.append(df_probid_user.iloc[i,16])
    syn.append(df_probid_user.iloc[i,17:-2].sum())
fig,a=plt.subplots(2,2,constrained_layout=True)
a[0][0].plot(re,label='Re')
a[0][0].set_title('RE')
a[0][1].plot(sem,label='SEM')
a[0][1].set_title('SEM')
a[1][0].plot(log,label='LOG')
a[1][0].set_title('LOG')
a[1][1].plot(syn,label='SYN')
a[1][1].set_title('SYN')
for ax in a.flat:
    ax.set(xlabel='Time', ylabel='No of errors')
print('Username: ',sel_user)
plt.show()    
    
```
